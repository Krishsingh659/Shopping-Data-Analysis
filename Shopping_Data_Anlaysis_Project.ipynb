```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df1=pd.read_csv("C:/Users/Krish/Downloads/shopping_data (2).csv")
df1
```

```python
x1=df1.iloc[:,3:]
x1
```

```python
plt.scatter(x1.iloc[:,0],x1.iloc[:,1])
plt.show()
```

```python
from sklearn.cluster import KMeans
k_means = KMeans(n_clusters=5)
df1["k_5"] = k_means.fit_predict(x1)
df1["k_5"]
```

```python
plt.scatter(x1.iloc[:,0],x1.iloc[:,1],c=df1["k_5"],cmap="rainbow")
plt.show()

from sklearn.metrics import silhouette_score
score=silhouette_score(x1,df1["k_5"])
print("Silhouette Score K Means: ",np.round(score,3))
```

```python
wcss=[]

for i in range(2,11):
    k_means = KMeans(n_clusters=i)
    k_means.fit(x1)
    wcss.append(k_means.inertia_) # inertia_ = WCSS 

plt.figure()
plt.scatter(range(2,11),wcss)
plt.plot(range(2,11),wcss,color="red")
plt.xlabel("No of clusters (K)")
plt.ylabel("WCSS")
plt.title("Elbow Method")
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df2 = pd.read_csv("C:/Users/Krish/Downloads/shopping_data (2).csv")
df2
```

```python
x2=df2.iloc[:,3:]
x2
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_sc=sc.fit_transform(x2)
x_sc
```

```python
from sklearn.cluster import dbscan
d_cluster = dbscan(x_sc,eps=0.5,min_samples=3)
d_cluster
```

```python
df2["cluster"]=d_cluster[1]
df2["cluster"]
```
